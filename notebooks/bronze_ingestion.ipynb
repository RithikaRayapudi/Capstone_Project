```python
#import libraries
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
#The path where raw stock CSV files are stored
data = Path("../data/raw/stocks")

# Reading individual stock datasets
aapl = pd.read_csv(data / "AAPL.csv")
msft = pd.read_csv(data / "MSFT.csv")
googl = pd.read_csv(data / "GOOGL.csv")

# Adding stock identifier column to each dataset
aapl["Stock"] = "AAPL"
msft["Stock"] = "MSFT"
googl["Stock"] = "GOOGL"

# Combining all stock datasets into a single DataFrame
stocks_df = pd.concat([aapl, msft, googl], ignore_index=True)

# Displaying first few rows
stocks_df.head()

```

```python
# Path to store Bronze output
bronze_output_path = Path("../data/processed")

# Write Bronze data to CSV
stocks_df.to_csv(
    bronze_output_path / "cleaned_stock_data.csv",
    index=False
)

# Display confirmation
print("Bronze data written to data/processed/cleaned_stock_data.csv")
```

```python
import pandas as pd
from pathlib import Path

# Load raw portfolio transactions
portfolio_path = Path("../data/raw/portfolio")
transactions_df = pd.read_csv(portfolio_path / "transactions.csv")

# Create processed directory if not exists
processed_path = Path("../data/processed")
processed_path.mkdir(parents=True, exist_ok=True)

# Write Bronze output
transactions_df.to_csv(
    processed_path / "cleaned_transactions.csv",
    index=False
)

transactions_df.head()
```
