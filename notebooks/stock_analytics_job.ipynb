```python
# Databricks Job Notebook
# This notebook is triggered by Airflow

from pyspark.sql.functions import avg

# Load analytics-ready stock data
df = spark.table("rithika.default.cleaned_stock_analytics")

# Simple aggregation to prove execution
result = df.groupBy("Stock").agg(
    avg("Daily_Return").alias("Avg_Daily_Return")
)

# Save result (overwrites on every run)
result.write.mode("overwrite").saveAsTable(
    "rithika.default.airflow_triggered_stock_summary"
)

print("Databricks analytics job executed successfully")
```
